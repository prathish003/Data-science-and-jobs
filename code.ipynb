```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset
df = pd.read_csv("jobs_in_data.csv")
# Display first few rows of the dataset
print(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
#Descriptive statistics
print()
print("Descriptive Statistics")
print(df.describe())
```

```python
#Correlation Analysis
# Select only numeric columns
numeric_df = df.select_dtypes(include='number')
# Compute correlation matrix
correlation_matrix = numeric_df.corr()
# Display correlation table
print("Correlation Table:")
print(correlation_matrix)
```

```python
#Categorical Graph
# Plot a pie chart for a categorical variable (e.g., Job Category)
plt.figure(figsize=(8, 8))
labels = df['job_category'].value_counts().index
sizes = df['job_category'].value_counts().values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Category')
plt.axis('equal')
plt.show()
```

```python
#Relational Graph
# Plot a line graph (e.g., Relationship between Years of Experience and Salary)
plt.plot(df['experience_level'], df['salary'], color='blue', marker='o')
plt.title('Relationship between Years of Experience and Salary')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
#Statistical graph
# Plot correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
